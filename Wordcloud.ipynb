```python
#import libraries
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
#open text file
file_content=open (f'blog_posts/BSR_ The regime and its enablers.txt', encoding="utf8").read()
```

```python
wordcloud = WordCloud(
                            stopwords=STOPWORDS,
                            background_color = 'white',
                            width=2000,
                            height=2000
                            ).generate(file_content)

plt.imshow(wordcloud)
plt.axis('off')
#plt.show()
plt.savefig('wc1.png')
```
